# Importing the libraries

```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Read, show, close the image

```python
img = cv2.imread("lena.jpg")
```

```python
def imshow(titl,image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(titl)
    plt.show()
```

```python
imshow("Show",img)
```

# Splitting and Merging the image

```python
b,g,r = cv2.split(img)
```

```python
zero = np.zeros(img.shape[:-1], dtype = "uint8")
```

```python
remake = cv2.merge([b,zero,zero])

imshow("Remade Blue",remake)
```

```python
remake = cv2.merge([zero,g,zero])
imshow("Remade Green",remake)
```

```python
remake = cv2.merge([zero,zero,r])
imshow("Remade Red",remake)

```

# Amplification

```python
amplified = cv2.merge([b+30,g-12,r+5])
imshow("Amplified", amplified)
```

# Translation of images

## Positioning

```python
width , height = img.shape[:-1]
```

```python
wdt, hgt = width/2, height/2
```

Creating a np float 

```python
position = np.float32([[1,0,wdt], [0,1,hgt]])
```

```python
positioned = cv2.warpAffine(img,position,(width,height))
```

```python
imshow("re-position", positioned)
```

## Rotating the picture

### using getRotationMatrix2d

```python
rotate = cv2.getRotationMatrix2D((wdt, hgt), 180 , 1)

imshow("Rotated", cv2.warpAffine(img, rotate, (width, height)))
```

### Using Transpose to Rotate

```python
transpose = cv2.transpose(img)

imshow("Rotated e transpose", transpose)
```

## Resizing of the Picture

```python
resized = cv2.resize(img , None ,fx = 0.75 , fy = 0.75)

imshow("resized", resized)
```

## Cropping the image

```python
start_row , start_col = int(height*0.45), int(width*0.45)

end_row, end_col = int(height*0.70) , int(width*0.70)
```

```python
imshow("Cropped", img[start_row:end_row,start_col:end_col])
```

# Blurring the image

```python
blurred = cv2.blur(img, (12,12))
```

```python
imshow("Blurred",blurred)
```

# Sharpening the image

```python
sharp_filter = np.array([[-1,-1,-1],
                         [-1,8,-1],
                         [-1,-1,-1]])
```

```python
edges = cv2.filter2D(img,-1,sharp_filter)
```

```python
imshow("Edges/ Feature Extraction", edges)
```
