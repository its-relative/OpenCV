# Lets close the image with the press of a click

```python
while True:
    cv2.imshow("Lena",img)

    if cv2.waitKey(1) & 0xff == ord("q"): # putting waitkey 0 won't work
        break # pressing q will close the window

cv2.destroyAllWindows()
```

# Lets remake the image somehow

```python
remake_img = cv2.merge([R,G,B])

plt.imshow(remake_img)
plt.show()
```

### lets see the difference it would create if we just put GBR

```python
remake_not_accept = cv2.merge([G,B,R])

plt.imshow(remake_not_accept)
plt.show()
```

```python
remake_not_accept_2 = cv2.merge([B,G,R])
plt.imshow(remake_not_accept)
plt.show()
```

#### so clearly the order we put those splitted dimensions matter in this case
