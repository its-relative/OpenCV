```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("lena.jpg")
```

```python
def imshow(titl,im):
    plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
    plt.title(titl)
    plt.show()
```

```python
imshow("Read",img)
```

```python
cv2.imshow("lena",img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

even tho it is closing with the click but for practice i'll write a code to close it through a click

```python
while True:
    cv2.imshow("Lena",img)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cv2.destroyAllWindows()
```

the only difference is that previously it was closing through any key but now only one

Translation

```python
width, height = img.shape[:-1]
```

```python
wdt, hgt = width/4 , height/4
```

```python
translated = np.float32([[1,0,wdt],[0,1,hgt]])

plt.imshow(cv2.warpAffine(img, translated, (width, height)))
plt.show()
```
